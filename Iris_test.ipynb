```python
import numpy as np
import pandas as pd

df_iris = pd.read_csv('/Users/andreytretyakov/IdeaProjects/JavaCourse/task_files/iris.data', 
                      names=['sepal_l', 'sepal_w', 'petal_l', 'petal_w', 'spec'])
df_iris.head()
```

```python
df_iris.info()
```

```python
avg_sizes = np.mean(df_iris.iloc[:, :-1])
print(f"Average sizes: { list(map(lambda x: '{:.2f}'.format(x), avg_sizes.values)) }")
```

```python
df_iris[df_iris.sepal_w < avg_sizes['sepal_w']]
```

```python
mean_versicolor = np.mean(df_iris[df_iris.spec == "Iris-versicolor"])
print(f"Average sizes of versicolors: { list(map(lambda x: '{:.2f}'.format(x), mean_versicolor.values)) }")
```

```python
df_iris[(df_iris.spec == "Iris-versicolor") & (df_iris.sepal_l < mean_versicolor['sepal_l'])]
```

```python
df_iris[df_iris.petal_l * df_iris.petal_w >= 5.0]
```

```python
df_iris[df_iris.petal_l * df_iris.petal_w < 2.0]
```

```python
df_iris[(df_iris['sepal_l'] == np.max(df_iris['sepal_l'][df_iris.spec == "Iris-setosa"])) &
       (df_iris.spec == "Iris-setosa")]
```
